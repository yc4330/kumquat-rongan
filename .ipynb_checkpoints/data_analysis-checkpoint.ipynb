```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f, ax=plt.subplots()
df = pd.read_csv("kumquat_variety.csv")
sns.boxplot(df, x="price", y="kind", hue="kind", width=.5, palette="RdBu")
sns.stripplot(df, x="price", y="kind",color=".3", size=6, jitter=False)
ax.set(ylabel="")
# sns.despine(trim=True, left=True)
```
