```python
# f, ax=plt.subplots()
df = pd.read_csv("variety.csv")

sns.set(rc={'figure.figsize':(6,4)})
sns.set(style="ticks")
sns.set_style("darkgrid")
# sns.boxplot(df, x="Price(¥/LB)", y="Variety", hue="Variety", width=.6, palette="RdYlGn")
sns.stripplot(df, x="Price(¥/LB)", y="Variety",color=".3", size=6, jitter=False).set_title("Rong'an Kumquat Price is Higher than Most of Other Places in China")
# filter Province="Rong'an", and stripplot
df_rongan = df[df['Province'] == "Rong'an"]
sns.stripplot(df_rongan, x="Price(¥/LB)", y="Variety", color="orange", size=15, marker="*", jitter=False)
plt.grid()
plt.show()
# sns.despine(trim=True, left=True)
# 
```
